```python
feature_names = [f"Feature {i+1}" for i in range(X.shape[1])]  # X is your feature matrix

for i, tree in enumerate(model.estimators_[:10]):
    decision_tree = tree[0]  # This is a DecisionTreeRegressor or DecisionTreeClassifier
    tree_ = decision_tree.tree_
    node_samples = tree_.n_node_samples
    node_impurity = tree_.impurity
    features_used = tree_.feature

    print(f"Tree {i+1}:")
    print("Number of Samples at Each Node:", node_samples)
    print("Impurity at Each Node:", node_impurity)
    print("Features Used at Each Node:", features_used)
    
    plt.figure(figsize=(20, 10))
    plot_tree(decision_tree, filled=True, feature_names=feature_names, class_names=model.classes_)
    plt.title(f"Decision Tree {i+1}")
    plt.show()
```

```python
feature_sample_counts = {}

for tree in model.estimators_:
    tree_ = tree[0].tree_  # Access the tree structure (tree_ contains information like features, impurity, etc.)
    node_samples = tree_.n_node_samples
    features_used = tree_.feature
    for feature_idx, sample_count in zip(features_used, node_samples):
        if feature_idx >= 0:  # If a valid feature was used (not a leaf node, which has feature_idx == -2)
            if feature_idx not in feature_sample_counts:
                feature_sample_counts[feature_idx] = 0
            feature_sample_counts[feature_idx] += sample_count

for feature_idx, total_samples in feature_sample_counts.items():
    print(f"Feature {feature_idx}: {total_samples} samples")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier

feature_accuracies = {}

for feature in results_df.columns[1:-2]:  # Exclude first column (if it's an index) and y_test, y_pred
    X_feature = results_df[[feature]].to_numpy()  # Use only one feature at a time

    # Split based on test indices from before
    X_feature_train, X_feature_test = X_feature[results_df.index.difference(test_indices)], X_feature[test_indices]
    y_feature_train, y_feature_test = y[results_df.index.difference(test_indices)], y_test  # Same y values

    # Train a classifier (you can change the model)
    model = GradientBoostingClassifier(random_state=42)
    model.fit(X_feature_train, y_feature_train)
    y_feature_pred = model.predict(X_feature_test)

    # Compute accuracy
    feature_accuracies[feature] = accuracy_score(y_feature_test, y_feature_pred)

# Convert to DataFrame for better visualization
import pandas as pd
accuracy_df = pd.DataFrame.from_dict(feature_accuracies, orient='index', columns=['Accuracy'])
print(accuracy_df.sort_values(by='Accuracy', ascending=False))
```

```python
# Convert feature accuracy dictionary to DataFrame
accuracy_df = pd.DataFrame.from_dict(feature_accuracies, orient='index', columns=['Accuracy'])
accuracy_df.index.name = 'feature'
accuracy_df.reset_index(inplace=True)

# Extract feature type
accuracy_df['feature_type'] = accuracy_df['feature'].apply(lambda x: x.split('_')[0])

# Aggregate accuracy by feature type (e.g., mean accuracy per type)
feature_type_accuracy = accuracy_df.groupby('feature_type')['Accuracy'].mean()

print(feature_type_accuracy.sort_values(ascending=False))
```
